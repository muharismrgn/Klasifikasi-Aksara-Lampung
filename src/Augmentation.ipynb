```python
import cv2
import os
import glob
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Tentukan direktori utama
main_directory = '../dataset/preprocessed'

# Konfigurasi augmentasi gambar
datagen = ImageDataGenerator(
    rotation_range=30,    # Rentang rotasi dalam derajat
    shear_range=0.2,      # Rentang pemutarannya
    zoom_range=0.2,       # Rentang zoom-in dan zoom-out
    horizontal_flip=True, # Melakukan flip horizontal
    vertical_flip=True,   # Melakukan flip vertikal
    fill_mode='nearest'   # Cara mengisi piksel yang kosong saat augmentasi
)

# List semua direktori label dalam direktori utama
label_dirs = [f.path for f in os.scandir(main_directory) if f.is_dir()]

# Loop melalui setiap direktori label
for directory in label_dirs:
    # Ambil semua path gambar dengan ekstensi .png dalam direktori label
    images = glob.glob(os.path.join(directory, '*.png'))
    
    # Loop melalui setiap gambar
    for image_path in images:
        print(image_path)
        # Baca gambar menggunakan OpenCV dan konversi warna ke RGB
        img = cv2.cvtColor(cv2.imread(image_path, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
        
        # Reshape gambar menjadi bentuk yang sesuai dengan generator
        image = img.reshape((1, ) + img.shape) 
        
        # Ekstrak nama file tanpa ekstensi dari path gambar
        base_name = os.path.basename(image_path)  # Misal: ah_01.png
        file_name_without_ext = os.path.splitext(base_name)[0]  # Misal: ah_01

        # Ekstrak folder name dari nama file (Misal: dari ah_01, folder name menjadi AH)
        folder_name = file_name_without_ext.split('_')[0].upper()
        
        j = 0
        # Tentukan prefix untuk nama file yang akan disimpan
        prefix = f'aug_{file_name_without_ext.split("_")[0]+file_name_without_ext.split("_")[1]}'
        total_augmentasi = 7  # Jumlah total augmentasi yang diinginkan
        # Lakukan augmentasi gambar
        for batch in datagen.flow(image, batch_size=16, 
                                  save_to_dir=os.path.join(main_directory, folder_name), 
                                  save_prefix=prefix, 
                                  save_format='png'):
            j += 1
            if j == total_augmentasi:
                break
```
